```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# 
img1 = mpimg.imread("figures/output1.png")
img2 = mpimg.imread("figures/output2.png")

# 
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))  # 12

# 
ax1.imshow(img1)
ax1.axis('off')  # 
ax1.set_title("Left Image")  # 

ax2.imshow(img2)
ax2.axis('off')  # 
ax2.set_title("Right Image")  # 

# 
plt.tight_layout()

# 
plt.show()

# 
fig.savefig("output_combined.png", bbox_inches='tight')
```
