```python
import numpy as np
import matplotlib.pyplot as plt

# 
plt.rcParams.update({
    "text.usetex": False,
    "font.family": "Times New Roman",
    "font.size": 12,
    "figure.figsize": (8, 6)
})

# 
def swish(x):
    return x / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x):
    return np.where(x < 0, 0.1 * x, x)

def tanh(x):
    return np.tanh(x)

def elu(x):
    return np.where(x < 0, np.exp(x) - 1, x)

# 
x = np.linspace(-5, 5, 200)
y_swish = swish(x)
y_relu = relu(x)
y_leaky_relu = leaky_relu(x)
y_tanh = tanh(x)
y_elu = elu(x)

# 
plt.plot(x, y_swish, label="Swish", color="green", linewidth=1.8)
plt.plot(x, y_relu, label="ReLU", color="red", linewidth=1.8)
plt.plot(x, y_leaky_relu, label="Leaky ReLU", color="orange", linewidth=1.8)
plt.plot(x, y_tanh, label="Tanh", color="blue", linewidth=1.8)
plt.plot(x, y_elu, label="ELU", color="purple", linewidth=1.8)

# 
plt.grid(color='lightgray', linestyle='--', linewidth=0.5)

# 
plt.title("Activation Functions", fontsize=18)
plt.xlabel("Input (x)", fontsize=14)
plt.ylabel("Output (y)", fontsize=14)

# 
plt.legend(fontsize=12, frameon=False)

# 
plt.ylim(-1, 2)

# 
plt.tight_layout()
plt.show()
```
